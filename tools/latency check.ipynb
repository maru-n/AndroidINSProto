```python
data = np.loadtxt("latency.data")
data.shape
```

```python
plt.figure(figsize=(20,5))
plt.plot(data)
plt.show()
plt.plot(data)
plt.xlim(0,3000)
plt.show()
```

```python
plt.hist(data, bins=1000)
av = np.average(data)
plt.title("average=%f"%av)
plt.show()
```

```python
d = data[data<0.025]
av = np.average(d)
plt.hist(d, bins=1000)
plt.title("average=%f"%av)
plt.xlim(av-0.001, av+0.001)
plt.show()
```
